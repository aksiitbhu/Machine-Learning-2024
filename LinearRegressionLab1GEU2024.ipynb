Error function for 1 degree polynomial with zero intercept
Equation of the line is â€‹y = mx + c
1. Slope of the line is 1 i.e m = 1.
2. Intercept is 0 i.e. c = 0.

```python
import numpy as np
import matplotlib.pyplot as plt
m = 1
x = np.linspace(1,10,11)
y = m*x
plt.title('Graph of equation y = x')
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
w1 = np.linspace(-6,9,20)
error = 0
it = len(x)
for i in range(it):
  error += (w1*x[i]- y[i])**2
plt.plot(w1,error)
plt.xlabel('W')
plt.ylabel('Error')
plt.title('W vs Error')
plt.scatter(w1,error,c='red')
plt.show()
```
